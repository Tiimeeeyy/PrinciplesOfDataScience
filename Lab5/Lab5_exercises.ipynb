## Standard Error
21. For each of the ten distributions that you visualized above, do the following:

    1. Calculate the sum of 100 observations from the distribution, a total of 100 times.
    2. Calculate the Standard Error and the Expected Value of the experiments
    3. Visualize the outcomes of the experiments and indicate the range between the expected value minus the standard error and the expected value plus the standard error.
    4. Determine how many outcomes lie within the indicated area

What do you notice? 

```python
import matplotlib.pyplot as plt
import numpy as np

# List of distributions
distributions = [obs1, obs2, obs3, obs4, obs5, obs6, obs7, obs8, obs9, obs10]

for i, dist in enumerate(distributions, 1):
    sums = [np.sum(dist.sample(100)) for _ in range(100)]
    EV = np.mean(sums)
    SE = np.std(sums)

    # Visualize the outcomes
    plt.figure(i)
    plt.hist(sums, bins=20, alpha=0.5)
    plt.axvline(EV - SE, color='r', linestyle='dashed', linewidth=2)
    plt.axvline(EV + SE, color='r', linestyle='dashed', linewidth=2)
    plt.title(f'Distribution {i}: EV={EV:.2f}, SE={SE:.2f}')

    # Determine how many outcomes lie within the indicated area
    within_range = np.sum((EV - SE <= sums) & (sums <= EV + SE))
    print(f'Distribution {i}: {within_range} outcomes lie within the indicated area')

plt.show()
    
```
